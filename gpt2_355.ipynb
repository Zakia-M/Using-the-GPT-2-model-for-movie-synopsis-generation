## Prepare data

```python
# Import libraries.
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# import data
data = pd.read_csv("datasets/movies_synopsis_cleaned.csv", header=0)
print("dataset shape : ", data.shape, "\n")
print("dataset columns", data.columns, "\n")
plot=data[['plot']]
# put the first column on a csv file to use it to finetune the model.
plot.to_csv ('plots.csv', index = False, header=True)
print("print the first plot : \n")
print(plot['plot'][0])
```

## gpt_2_simple 355

### Download and install the model

```python
%tensorflow_version 1.x
!pip install -q gpt-2-simple
import gpt_2_simple as gpt2
from datetime import datetime
from google.colab import files
```

```python
# gpt2.download_gpt2(model_name="124M")
gpt2.download_gpt2(model_name="355M")
```

### Finetune the model

```python
sess = gpt2.start_tf_sess()
plots="plots.csv"
gpt2.finetune(sess,
              dataset=plots,
              model_name='355M',
              steps=2000,       # ~~ epoches
              restore_from='fresh',
              run_name='run1',
              print_every=50,   # print the loss
              sample_every=100, # print a sample
              save_every=500
              )
```

### Generate synopsis

```python
test = gpt2.generate(sess,
              temperature = 1.0,
              nsamples=50,
              top_k=40,
              batch_size=5,
              prefix = '<|startoftext|>', 
              truncate='<|endoftext|>',
              return_as_list = True
              )

for ix, i in enumerate(test):
  print("movie overview {}:".format(ix))
  print(i)
  print()
```
